### Install and Import modules

```python
#!python -m pip install spacy
#!python -m pip install wordcloud
#!python -m pip install textacy
import matplotlib as mpl
from wordcloud import WordCloud, STOPWORDS
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import matplotlib.pyplot as plt
import re, string, sys, warnings, random
import pandas as pd
import numpy as np
%matplotlib inline
import seaborn as sns
warnings.filterwarnings('ignore')
from textacy.text_stats import TextStats
from utils import *
```

### load messages and prepare

```python
msg_file = '../../data/messages_computational_analysis.csv'
all_msg_df  = pd.read_csv(msg_file)

human_msgs = all_msg_df[all_msg_df['Source'] == "HUM"]["Message Text"]#.values
print(human_msgs[:3])
print('---')
ai_msgs = all_msg_df[all_msg_df['Source'] == "AI"]["Message Text"]#.values
print(ai_msgs[:3])

print(len(all_msg_df))

human_msgs    = pd.DataFrame(human_msgs[:].apply(lambda x: clean_text(x)))
ai_msgs       = pd.DataFrame(ai_msgs[:].apply(lambda x: clean_text(x)))
print(len(human_msgs))
print(len(ai_msgs))

all_msg_df.head()
```

### Plot Wordclouds

```python
mpl.rcParams['figure.figsize']= (8.0,8.0)  
mpl.rcParams['font.size']     = 12            
mpl.rcParams['savefig.dpi']   = 100             
mpl.rcParams['figure.subplot.bottom']=.1 

my_stopwords = list(STOPWORDS)
my_stopwords.remove("don't")
my_stopwords.append("ve")
my_stopwords.append("tha")

wordcloud = WordCloud(    background_color='white',
                          stopwords=my_stopwords,
                          max_words=300,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(human_msgs));
print(wordcloud);
fig = plt.figure(1);
plt.imshow(wordcloud.recolor(color_func=red_color_func, random_state=42));
plt.axis('off');
plt.show();

wordcloud2 = WordCloud(   background_color='white',
                          stopwords=my_stopwords,
                          max_words=300,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(ai_msgs));
print(wordcloud2);
fig = plt.figure(1);
plt.imshow(wordcloud2.recolor(color_func=blue_color_func, random_state=42));
plt.axis('off');
plt.show();
```

### Uni and Ngrams (3-grams)

```python
aigen_unigram
```

```python
common_words = get_top_n_gram(human_msgs["Message Text"], ngram = 1, n=6, stopwords = 'english')
human_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

common_words = get_top_n_gram(ai_msgs["Message Text"], ngram = 1, n=6, stopwords = 'english')
aigen_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

#print(human_unigram[:5])
#print('---')
#print(aigen_unigram[:5])

fig, ax1 = plt.subplots(figsize = (4,2))
human_unigram.sort_values(by='count').plot(kind='barh', color = 'salmon', ax = ax1);
plt.yticks(np.arange(6), human_unigram.sort_values(by='count')['unigram']);
plt.xticks(np.arange(0,16,2));
plt.title("Human Messages Unigrams")
#plt.show()
fig, ax2 = plt.subplots(figsize = (4,2))
aigen_unigram.sort_values(by='count').plot(kind='barh', color = 'dodgerblue', ax = ax2);
plt.yticks(np.arange(6), aigen_unigram.sort_values(by='count')['unigram']);
plt.xticks(np.arange(0,16,2));
plt.title("AI Messages Unigrams")
plt.show()
```

```python
common_words = get_top_n_gram(human_msgs["Message Text"], ngram = 3, n=6, stopwords = None)
human_ngram = pd.DataFrame(common_words, columns = ['ngram' , 'count'])
#print(human_ngram[:5])

common_words = get_top_n_gram(ai_msgs["Message Text"], ngram = 3, n=6, stopwords = None)
aigen_ngram = pd.DataFrame(common_words, columns = ['ngram' , 'count'])
#print(aigen_ngram[:5])

#plt.figure(figsize = (12,2))
fig, ax1 = plt.subplots(figsize = (4,2))
human_ngram.sort_values(by='count').plot(kind='barh', color = 'salmon', ax = ax1);
plt.yticks(np.arange(6), human_ngram.sort_values(by='count')['ngram']);
plt.title("Human Messages: Trigrams");


fig, ax2 = plt.subplots(figsize = (4,2))
aigen_ngram.sort_values(by='count').plot(kind='barh', color = 'dodgerblue', ax = ax2);
plt.yticks(np.arange(6), aigen_ngram.sort_values(by='count')['ngram']);
plt.title("AI Messages: Trigrams");

#ax2 = plt.subplot(1,2,2)
#aigen_ngram.plot(kind='barh', color = 'dodgerblue', ax = ax2);
#plt.yticks(np.arange(6), aigen_ngram['ngram']);
##plt.title("GPT4 Messages: Trigrams")
#plt.show()
```
