## Computational Analyses III: Semantic Similarity

Sentence Embeddings using Siamese BERT-Networks: The Sentence Transformer library is available on [pypi](https://pypi.org/project/sentence-transformers/) and [github](https://github.com/UKPLab/sentence-transformers). The library implements code from the ACL 2019 paper entitled "[Sentence-BERT: Sentence Embeddings using Siamese BERT-Networks](https://www.aclweb.org/anthology/D19-1410.pdf)" by Nils Reimers and Iryna Gurevych.

### Install and Import modules

```python
#!conda install -c conda-forge sentence-transformers -y  # install module
from sentence_transformers import SentenceTransformer
from sentence_transformers.util import pytorch_cos_sim
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import nilearn.connectome
import seaborn as sns
from scipy import stats

# Load the BERT model. Various models trained on Natural Language Inference (NLI) https://github.com/UKPLab/sentence-transformers/blob/master/docs/pretrained-models/nli-models.md and 
# Semantic Textual Similarity are available https://github.com/UKPLab/sentence-transformers/blob/master/docs/pretrained-models/sts-models.md
model = SentenceTransformer('stsb-mpnet-base-v2')
```

### Set up example corpus, compute similarity, and plot

```python
# set up sentences and transform into embeddings
sentences_demo = [  'I like my phone.',
                'Your cellphone looks great.',
                'Will it snow tomorrow?',
                'Hurricanes have hit the US.',
                'How old are you?',
                'What is your age?',
                'You are powerful, elegant, and in control. Do not let a text disrupt that grace on the road. Drive now, chat later',
                'The UAE celebrates our strength and progress. Let us champion safety too. Prioritize life over a text.',
                'Striding forward with style and purpose? The road is your runway. Keep it classy and safe. Avoid texting while driving.',
                'Your journey, your rules. But remember, in the dance of life, safety always takes the lead. Stay focused.',
                'The story of UAE women is one of ambition and inspiration. Do not let a text interrupt your legacy. Your message and future can wait']
sentence_embeddings_demo = model.encode(sentences_demo)

# compute similarity
matcos_demo = np.zeros((len(sentences_demo),len(sentences_demo)))
for i in range(len(sentences_demo)):
  for j in range (len(sentences_demo)):
    matcos_demo[i,j] = pytorch_cos_sim(sentence_embeddings_demo[i,:], sentence_embeddings_demo[j,:])

# plot
fig, ax = plt.subplots()
im = ax.imshow(matcos_demo, cmap = 'YlOrRd', vmin = 0, vmax = 1);
ax.set_yticks(np.arange(len(sentences_demo)))
ax.set_yticklabels(sentences_demo)
fig.colorbar(im, ax=ax)
plt.show()
```

```python
# hypothetical matrix if similarity were fully defined by source
hypo = np.zeros((30,30))
hypo[:15,:15] = 1
hypo[15:,15:] = 1
plt.imshow(hypo, cmap = 'YlOrRd', vmin = 0, vmax = 1);
plt.colorbar();
```

```python
# now real message-to-message similarity

mat = np.corrcoef(sentence_embeddings)
m1 = mat[:30, :30]
m1.shape

matcos = np.zeros((30,30))
for i in range(30):
  for j in range (30):
    matcos[i,j] = pytorch_cos_sim(sentence_embeddings[i,:], sentence_embeddings[j,:])

plt.imshow(matcos, cmap = 'YlOrRd', vmin = 0, vmax = 1);
plt.colorbar()
#plt.show()

#plt.imshow(np.corrcoef(sentence_embeddings)[:, :], cmap = 'YlOrRd', vmin = 0, vmax = 1);
#plt.colorbar()

sns.despine(left=True, bottom=True)
plt.xticks([])
plt.yticks([])

fig, ax = plt.subplots(figsize = (15,15))
im = ax.imshow(matcos, cmap = 'YlOrRd', vmin = 0, vmax = 1);
ax.set_yticks(np.arange(len(sentences)))
ax.set_yticklabels(sentences, size = 18)
fig.colorbar(im, ax=ax)
plt.show()
```

```python
# just quick test -- standard corr and cos are the same . ;-)
cos = nilearn.connectome.sym_matrix_to_vec(matcos, discard_diagonal=True)
cor = nilearn.connectome.sym_matrix_to_vec(mat, discard_diagonal=True)
np.corrcoef(cos, cor)
```

```python
# separate ai and human-generated messages and compare in terms of similarity: within-AI, within-Human, and ai-to-human
aigens  = nilearn.connectome.sym_matrix_to_vec(matcos[:15, :15], discard_diagonal=True)
humans  = nilearn.connectome.sym_matrix_to_vec(matcos[15:, 15:], discard_diagonal=True)
aitohum = nilearn.connectome.sym_matrix_to_vec(matcos[15:, :15], discard_diagonal=True)

print(np.mean(aigens))
print(np.mean(humans))
print(np.mean(aitohum))
print('---')

aimeans       =  np.nanmean(matcos[:15, :15], axis = 0)
hummeans      =  np.nanmean(matcos[15:, 15:], axis = 0)
aitohummeans  =  np.nanmean(matcos[15:, :15], axis = 0)

print(np.mean(aimeans))
print(np.mean(hummeans))
print(np.mean(aitohummeans))
print('---')

print(stats.ttest_ind(aimeans, hummeans))
print(stats.ttest_ind(aimeans, aitohummeans))
print(stats.ttest_ind(aitohummeans, hummeans))
```
