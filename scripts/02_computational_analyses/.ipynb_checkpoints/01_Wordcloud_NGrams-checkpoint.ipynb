### Read tweets by humans (scraped) and Bloom, all files and selection

```python
msg_file = '../../data/messages_computational_analysis.csv'
all_msg_df  = pd.read_csv(msg_file)

human_msgs = all_msg_df[all_msg_df['Source'] == "HUM"]["Message Text"]#.values
print(human_msgs[:3])
print('---')
ai_msgs = all_msg_df[all_msg_df['Source'] == "AI"]["Message Text"]#.values
print(ai_msgs[:3])

print(len(all_msg_df))
all_msg_df.head()
```

```python
def clean_text(text):
   text = text.lower()                                               # make text lowercase  
   text = re.sub(r'\[.*?\]', '', text)                               # remove text in square brackets
   text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)  # remove punctuation   
   text = re.sub(r'\w*\d\w*', '', text)                              # remove words containing numbers
   return text

human_msgs    = pd.DataFrame(human_msgs[:].apply(lambda x: clean_text(x)))
ai_msgs       = pd.DataFrame(ai_msgs[:].apply(lambda x: clean_text(x)))
print(len(human_msgs))
print(len(ai_msgs))
```

```python
mpl.rcParams['figure.figsize']= (8.0,8.0)  
mpl.rcParams['font.size']     = 12            
mpl.rcParams['savefig.dpi']   = 100             
mpl.rcParams['figure.subplot.bottom']=.1 

my_stopwords = list(STOPWORDS)
my_stopwords.remove("don't")
my_stopwords.append("ve")
my_stopwords.append("tha")


def blue_color_func(word, font_size, position, orientation, random_state=None,**kwargs):
    return "dodgerblue" #hsl(228, 49%, 56%)" 

def red_color_func(word, font_size, position, orientation, random_state=None,**kwargs):
    return "salmon"  #hsl(0, 49%, 65%)" 

wordcloud = WordCloud(    background_color='white',
                          stopwords=my_stopwords,
                          max_words=300,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(human_msgs))
print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud.recolor(color_func=red_color_func, random_state=42))
plt.axis('off')
plt.show();

wordcloud2 = WordCloud(   background_color='white',
                          stopwords=my_stopwords,
                          max_words=300,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(ai_msgs))
print(wordcloud2)
fig = plt.figure(1)
plt.imshow(wordcloud2.recolor(color_func=blue_color_func, random_state=42))
plt.axis('off')
plt.show();
```

### Uni and Bigrams

```python
def get_top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_words(human_msgs["Message Text"], 10)
human_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

common_words = get_top_n_words(ai_msgs["Message Text"], 10)
aigen_unigram = pd.DataFrame(common_words, columns = ['unigram' , 'count'])

#print(human_unigram[:5])
#print('---')
#print(aigen_unigram[:5])

plt.figure(figsize = (10,2))
ax1 = plt.subplot(1,2,1)
human_unigram.plot(kind='bar', color = 'salmon', ax = ax1);
plt.xticks(np.arange(10), human_unigram['unigram'], rotation=90);
plt.title("Human Messages Unigrams")
#plt.show()
ax2 = plt.subplot(1,2,2)
aigen_unigram.plot(kind='bar', color = 'dodgerblue', ax = ax2);
plt.xticks(np.arange(10), aigen_unigram['unigram'], rotation=90);
plt.title("GPT4 Messages Unigrams")
plt.show()
```

```python
def get_top_n_gram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2,2),stop_words='english')
    vec.fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in      vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]

common_words = get_top_n_gram(human_msgs["Message Text"], 10)
human_bigram = pd.DataFrame(common_words, columns = ['bigram' , 'count'])
#print(human_bigram[:5])

common_words = get_top_n_gram(ai_msgs["Message Text"], 10)
aigen_bigram = pd.DataFrame(common_words, columns = ['bigram' , 'count'])
#print(aigen_bigram[:5])

plt.figure(figsize = (10,2))
ax1 = plt.subplot(1,2,1)
human_bigram.plot(kind='bar', color = 'salmon', ax = ax1);
plt.xticks(np.arange(10), human_bigram['bigram'], rotation=90);
plt.title("Human Messages Bigrams")

ax2 = plt.subplot(1,2,2)
aigen_bigram.plot(kind='bar', color = 'dodgerblue', ax = ax2);
plt.xticks(np.arange(10), aigen_bigram['bigram'], rotation=90);
plt.title("GPT4 Messages Bigrams")
plt.show()
```
