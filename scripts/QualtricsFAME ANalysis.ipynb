```python
import sys, os
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
%matplotlib inline
import seaborn as sns
sns.set_theme(style="whitegrid")
import numpy as np
```

```python
df = pd.read_csv('../../privatedata/data4.csv')#, skiprows = [1,2])
df.iloc[:20, :20]
```

```python

print(len(df))

df = df[df.Progress == 100]
print(len(df))

df = df[df['Duration (in seconds)'] > 240]
print(len(df))
```

```python
df.to_csv('../../privatedata/data4a.csv')
```

```python
df.describe()
```

```python
df.groupby(['Q3']).describe()
```

```python
df_quality_human = df.filter(regex = 'Quality_Human',axis=1)
df_quality_ai    = df.filter(regex = 'Quality_AI',axis=1)

df_quality_ai.mean().hist(    alpha    = 0.7, label = 'Quality AI');
df_quality_human.mean().hist( alpha = 0.7, label = 'Quality Human');
plt.xlim(1, 5);
plt.legend()
plt.show()

# invert because of item polarity (low scores - high quality)
quality_ai    = 6 - df_quality_ai.mean().values
quality_human = 6 - df_quality_human.mean().values

print(stats.ttest_ind(quality_ai, quality_human))
print(np.mean(quality_human))
print(np.mean(quality_ai))


# for clarity, high scores already mean higher clarity
df_clarity_human = df.filter(regex='Clarity_Human', axis=1)
df_clarity_ai    = df.filter(regex='Clarity_AI',    axis=1)

df_clarity_ai.mean().hist(    alpha = 0.7, label = 'Clarity AI');
df_clarity_human.mean().hist( alpha = 0.7, label = 'Clarity Human');
plt.xlim(1,5);
plt.legend()

clarity_ai    = df_clarity_ai.mean().values
clarity_human = df_clarity_human.mean().values

print(stats.ttest_ind(clarity_ai, clarity_human))
print(np.mean(clarity_human))
print(np.mean(clarity_ai))
```

```python
dfh             = pd.DataFrame()
dfh['rating']   = clarity_human
dfh['type']     = 'Human'
dfh['question'] = 'Clarity'

dfa             = pd.DataFrame()
dfa['rating']   = clarity_ai
dfa['type']     = 'AI'
dfa['question'] = 'Clarity'

df_clarity      = dfh.append(dfa)

####

dfh             = pd.DataFrame()
dfh['rating']   = quality_human
dfh['type']     = 'Human'
dfh['question'] = 'Quality'

dfa             = pd.DataFrame()
dfa['rating']   = quality_ai
dfa['type']     = 'AI'
dfa['question'] = 'Quality'

df_quality      = dfh.append(dfa)

###

df_all = df_clarity.append(df_quality)
df_all.head()
```

```python
sns.set_style("ticks")


plt.figure(figsize = (4,6))

ax = sns.violinplot(data=df_all, x="question", y="rating", hue="type", 
               split=True, inner="quart", linewidth=1,
               palette={"Human": "r", "AI": "b"});
for violin, alpha in zip(ax.collections[::], [0.3,0.3, 0.3, 0.3]):
    violin.set_alpha(alpha)

sns.swarmplot(data=df_all, 
              x="question", 
              y="rating", 
              hue="type",
              palette={"Human": "r", "AI": "b"});

#plt.legend(bbox_to_anchor=(1.0, 1), borderaxespad=0);
#plt.legend(loc = 2);
plt.legend([],[], frameon=False)
plt.ylim(1,5);

sns.despine(offset= 10, trim=True);
plt.yticks(np.arange(1,6))


plt.savefig('fig2a.svg')
#sns.despine(left=True)
```
