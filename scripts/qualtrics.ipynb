```python
import sys, os
import matplotlib.pyplot as plt
from scipy import stats
import pandas as pd
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../data/uae_survey_cleaned_reversed_reduced_avg.csv')#, skiprows = [1,2])

print(len(df))
df.head()
```

```python
df.describe()
```

```python
df.filter(regex = 'clarity_ai').shape
```

```python
df_quality_human = df.filter(regex = 'quality_hum',axis=1)
df_quality_ai    = df.filter(regex = 'quality_ai',axis=1)

sns.set_theme(style="ticks")

plt.figure(figsize = (4,3))
df_quality_ai.mean().hist(    alpha    = 0.7, label = 'Quality AI');
df_quality_human.mean().hist( alpha = 0.7, label = 'Quality Human');
plt.xlim(1, 5);
plt.legend()
sns.despine()
plt.show()
```

```python

# for clarity, high scores already mean higher clarity
df_clarity_human = df.filter(regex='clarity_hum', axis=1)
df_clarity_ai    = df.filter(regex='clarity_ai',    axis=1)

plt.figure(figsize = (4,3))
df_clarity_ai.mean().hist(    alpha = 0.7, label = 'Clarity AI');
df_clarity_human.mean().hist( alpha = 0.7, label = 'Clarity Human');
plt.xlim(1,5);
plt.legend()
```

```python
df_clarity_human.shape
```

```python
clarity_human = df_clarity_human.mean().values
print(len(clarity_human))
clarity_ai = df_clarity_ai.mean().values

quality_human = df_quality_human.mean().values
quality_ai = df_quality_ai.mean().values
```

```python
dfh             = pd.DataFrame()
dfh['rating']   = clarity_human
dfh['type']     = 'Human'
dfh['question'] = 'Clarity'

dfa             = pd.DataFrame()
dfa['rating']   = clarity_ai
dfa['type']     = 'AI'
dfa['question'] = 'Clarity'

df_clarity = pd.concat([dfh, dfa], ignore_index = True)

####

dfh             = pd.DataFrame()
dfh['rating']   = quality_human
dfh['type']     = 'Human'
dfh['question'] = 'Quality'

dfa             = pd.DataFrame()
dfa['rating']   = quality_ai
dfa['type']     = 'AI'
dfa['question'] = 'Quality'

df_quality = pd.concat([dfh, dfa], ignore_index = True)

###

df_all  = pd.concat([df_clarity, df_quality], ignore_index = True)
df_all
```

```python
sns.set_style("ticks")


plt.figure(figsize = (4,6))

ax = sns.violinplot(data=df_all, x="question", y="rating", hue="type", 
               split=True, inner="quart", linewidth=1,
               palette={"Human": "r", "AI": "b"});
for violin, alpha in zip(ax.collections[::], [0.3,0.3, 0.3, 0.3]):
    violin.set_alpha(alpha)

sns.swarmplot(data=df_all, 
              x="question", 
              y="rating", 
              hue="type",
              palette={"Human": "r", "AI": "b"});

#plt.legend(bbox_to_anchor=(1.0, 1), borderaxespad=0);
plt.legend(loc = 3);
#plt.legend([],[], frameon=False)
plt.ylim(1,5);

sns.despine(offset= 10, trim=True);
plt.yticks(np.arange(1,6))


plt.savefig('fig2a.svg')
#sns.despine(left=True)
```
